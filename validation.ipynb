```python
#label feature size 28 * 28
import torchvision.models as models
import torch.nn as nn
import torch
import numpy as np
from resnet50 import resnet50
import operators as op_nn


X = np.random.random(size=(1,3,112,112)) 
resnet = resnet50()
resnet.load_state_dict()
X_list= X.tolist()
```

```python
class Resnet50(nn.Module):
  def __init__ (self,input_size=1):
    super(Resnet50, self).__init__()
    self.model = models.resnet50(pretrained=True)
  def forward(self, x):
    return self.model(x)
```

```python
X_tensor = torch.tensor(X,dtype=torch.float32)
model = Resnet50().to("cpu")

model_list =[]
#train_dataLoader ,test_dataloader는 위에서 정의함.
model.eval()
with torch.no_grad():
    pred_y = model(X_tensor)
    model_list.append(pred_y)
```

```python
savings = []
count =0
for i in range(32):
    softmax__torch_y= torch.softmax(model_list[0][i],dim=0)
    if torch.argmax(softmax_list[i]) == torch.argmax(softmax__torch_y):
        count+=1
    savings.append((torch.argmax(softmax_list[i]),torch.argmax(softmax__torch_y)))

```
